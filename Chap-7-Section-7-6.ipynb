# Bilans macroscopiques :

Par des bilans macroscopiques sur un système fluidique pouvant comprendre plusieurs entrées et plusieurs sorties, on obtient, ** en état de régime permanent** :

>## 1. Bilan de masse

$$\boxed { \sum_{entree} w_1 - \sum_{sortie} w_2 =0}(1)$$
$$\boxed { \sum_{entree} \rho_1 v_1 S_1 - \sum_{sortie} \rho_2 v_2 S_2 =0} (2) $$

Les vitesses sont les vitesses moyennes à travers ces surfaces


>## 2. Bilan de quantité de mouvement linéaire

$$\boxed{ \sum_{entree} ( v_1 w_1 + p_1 S_1) \vec{u_1}-\sum_{sortie}( v_2 w_2 + p_2 S_2) \vec{u_2} + m_{tot} \vec g=\vec F_{fluide \rightarrow surface}} (3)$$ 

Les vecteurs $\vec{u_1}$ et $\vec{u_2}$ sont les vecteurs unitaires dans la direction de l'écoulement à la surface. Les débits massiques $\vec{w_1}$ et $\vec{w_2}$ sont définis comme dans le bilan de masse.


>## 3. Bilan de quantité de mouvement angulaire

$$\boxed{ \sum_{entree} ( v_1w_1 + p_1 S_1) (\vec{r_1} \times \vec{u_1})
         -\sum_{sortie} ( v_2w_2 + p_2 S_2) (\vec{r_2} \times \vec{u_2}) 
         +\vec T_{ext}=
          \vec T_{fluide \rightarrow surface}} (4)$$



>## 4. Bilan d'énergie mécanique

$$\boxed{ \sum_{entree}  ( \frac {1}{2} v_1^2 + g h_1 + \frac {p_1}{\rho_1} )w_1-
          \sum_{sortie}  ( \frac {1}{2} v_2^2 + g h_2 + \frac {p_2}{\rho_2} ) w_2=-W_m +E_c+E_v} (5) $$


***

Les deux termes de sommation à gauche de l'équation représentent respectivement les entrées et les sorties du système fluidique qui n'est pas nécessairement constitué d'une seule conduite d'entrée et de sortie.

Deux outils sont nécessaires pour compléter l'équation d'énergie mécanique en particulier. Les termes à la gauche de cette équation constituent la célèbre équation de Bernouilli, proposée en 1738 par Daniel Bernouilli dans son livre Hydrodynamica:

>## Bilan d'énergie mécanique pour un fluide idéal sans travail mécanique (Bernouilli)

$$
\boxed
 { 
 \sum_{entree} ( \frac {1}{2} v_1^2 + g h_1 + \frac {p_1}{\rho_1} ) w_1
 - 
 \sum_{sortie} ( \frac {1}{2} v_2^2 + g h_2 + \frac {p_2}{\rho_2} ) w_2
 =0
 }
  (6)$$

Elle est valide pour un fluide idéal (sans friction), sans présence de pompe ou d'effets de compressibilité.

Pour un fluide réel, le terme $W_m$ représente le travail mécanique, par exemple une pompe, le terme  $E_c$ le travail de compression, qui est nul si le fluide est incompressible, et finalement le terme  $E_v$ qui représente le travail perdu par la friction du fluide sur les parois. 

Rappelons-nous du résultat obtenu au chapitre 6 en définissant le facteur de friction :

$$
\boxed
  { f =  \frac {1}{4}
         \frac {D}{L}
         \frac {\mathscr{P}_0 - \mathscr{P}_L} {\frac {1}{2} \rho { V }^2}
  } 
  (7)$$



et l'équation de Blasius qui peut être utilisée dans une gamme de nombres de Reynolds pour un écoulement turbulent dans une conduite lisse:

$$
\boxed {f = \frac {0.0791}{Re^{0.25}}} (8)
$$


###  Calcul des pertes d'énergie mécaniques par friction $E_v$

#### * pertes régulières

La friction sur les parois de la conduite demande un travail qui peut être calculé en utilisant la fonction de pertes d'énergie par friction $\Phi$ qui a été utilisée au chapitre 10 (section 10.4). Le résultat de l'intégration de ce terme est donné pour une conduite circulaire par :

$$
\boxed {E_v = 2 w\frac  {L}{D} V^2 f} (9)
$$

ou w est le débit massique. Pour une conduite non-circulaire on aurait:

$$
\boxed {E_v = \frac {1}{2} w\frac  {L}{R_h} V^2 f}
(10)$$

avec $R_h$ le rayon hydraulique. On utilise aussi souvent l'expression de friction ci-haut comme:

$$
\boxed {\hat E_v = 2 \frac  {L}{D} { V }^2 f
        \quad ,et \quad {\hat E_v =\frac {E_v}{w}} }
(11),(12)$$


#### * pertes singulières
Ce sont les pertes associées aux perturbations de l'écoulement comme les coudes, les changements de diamètres, la présence d'obstructions partielles comme des vannes, etc... On calcule ces dernières comme ci-haut, avec un coefficient de pertes associé à la perturbation identifiée, comme donné au tableau 7.5-1:

$$
\boxed {\hat E_v = \frac {1}{2} V^2 e_v}
(13)$$


>### Forme finale de l'équation de l'énergie mécanique:
En substituant les méthodes de calcul de $E_v$ vues ci-haut:

$$
  \boxed{ \sum_{entree}  ( \frac {1}{2} v_1^2 + g h_1 + \frac {p_1}{\rho_1} )w_1-
          \sum_{sortie}  ( \frac {1}{2} v_2^2 + g h_2 + \frac {p_2}{\rho_2}  )w_2=
          -W_m +E_c
          + 2 \sum_{i-conduites} ( w_i  \frac  {L}{D} V^2 f)_i 
          + \frac  {1}{2}\sum_{j-perturbations} (w_j  V^2 e_v)_j}
  (14)$$

ou les deux termes de sommation à droite représentent la sommation des pertes par friction dans toutes les longueurs de conduites i et la sommation des pertes singulières dans toutes les perturbations j.

## Exemple d'application de l'équation d'énergie mécanique
Une conduite de 5 cm de diamètre de 10 mètres de longueur transporte 300 lpm d'eau , elle se sépare alors en 3 conduites de 2, 3 et 4 cm qui font respectivement 20, 30 et 40 mètres de longueur et qui sont ensuite déchargées à l'atmosphère. Quels seront les débits dans les 3 conduites et quelle est la pression à la tête de la conduite de 5 cm? Toutes les conduites sont horizontales.
<img src='http://pierreproulx.espaceweb.usherbrooke.ca/images/Chap-7-1.png'>

#### Solution

Appliquons l'équation (14)

$$
  \boxed{ \sum_{entree}  ( \frac {1}{2} v_1^2 + g h_1 + \frac {p_1}{\rho_1} )w_1-
          \sum_{sortie}  ( \frac {1}{2} v_2^2 + g h_2 + \frac {p_2}{\rho_2}  )w_2=
          -W_m +E_c
          + \sum_{i-conduites} (w_i 2 \frac  {L}{D} V^2 f)_i 
          + \sum_{j-perturbations} (w_i \frac  {1}{2} V^2 e_v)_j}
$$

avec le point 1 qui se situe à l'entrée de la conduite de 5 cm et le point 2 qui est la décharge des 2 conduites de 2, 3 et 4 cm. 

Avec la densité constante, l'effet de la gravité nulle et la pression à la sortie des 3 conduites prises à 0, la pression atmosphérique. De plus, on notera $v_{21}$, $v_{22}$, et $v_{23}$ les vitesses dans les 3 conduites de sortie, et ainsi les débits seront aussi notés de la même façon. Les diamètres et longueurs des conduites utiliseront les mêmes notations. Dans ce cas, nous considérons que les pertes singulières sont négligeables.

$$
  \boxed{   ( \frac {1}{2} v_1^2 + \frac {p_1}{\rho} )w_1
           - ( \frac {1}{2} v_{21}^2  )w_{21}
           - ( \frac {1}{2} v_{22}^2  )w_{22}
           - ( \frac {1}{2} v_{23}^2  )w_{23}
           =2  (
           (w_1     \frac  {L_1}  {D_1}    V_1   ^2 f_1) +
           (w_{21}  \frac  {L_{21}}{D_{21}} V_{21}^2 f_{21}) +
           (w_{22}  \frac  {L_{22}}{D_{22}} V_{22}^2 f_{22}) +
           (w_{23}  \frac  {L_{23}}{D_{23}} V_{23}^2 f_{23}) 
            )}
  (15)$$

Puisque la vitesse et le débit sont directement reliés, cette équation contient 4 variables inconnues, les vitesses dans les 3 conduites après la séparation, plus la pression à la tête de la conduite de 5 cm. On a alors 1 équation et 4 inconnues! Même si on ajoute l'équation de conservation de la masse qui nous donne que la somme des 3 débits dans les conduites après séparation est égal au débit de la première conduite, on aura encore un nombre insuffisant d'équations pour résoudre.

Il faut alors décomposer le problème au lieu de l'affronter directement avec l'équation générale sur l'ensemble du système. Ainsi, en faisant un bilan à partir du point de séparation pour la première des 3 conduites:

$$
  \boxed{   ( \frac {1}{2} V_{21}^2 + \frac {p_{21}}{\rho}  )w_{21}
           - ( \frac {1}{2} V_{21}^2  )w_{21}
           =2 
           w_{21}  \frac  {L_{21}}{D_{21}} V_{21}^2 f_{21} 
           }
  (16)$$

ou en divisant par le débit

$$
  \boxed{   ( \frac {1}{2} V_{21}^2 + \frac {p_{21}}{\rho}  )
           - ( \frac {1}{2} V_{21}^2  )
           =2 
            \frac  {L_{21}}{D_{21}} V_{21}^2 f_{21} 
            }
(17)$$

qui peut être encore simplifiée en :

$$
  \boxed{   (  \frac {p_{21}}{\rho}  )
           =2 
            \frac  {L_{21}}{D_{21}}V_{21}^2 f_{21} 
           }
  (18)$$

de la même façon pour les 2 autres conduites:

$$
  \boxed{   (  \frac {p_{22}}{\rho}  )
           =2 
            \frac  {L_{22}}{D_{22}}V_{22}^2 f_{22} 
           }
  (19)$$

$$
  \boxed{   (  \frac {p_{23}}{\rho}  )
           =2 
            \frac  {L_{23}}{D_{23}}V_{23}^2 f_{23} 
            }
  (20)$$

Il faut voir ici que la pression $P_{21}$ est identique à la pression $P_{22}$ et $P_{23}$ car ces 3 pressions viennent du même point de séparation de la conduite de 5 cm en 3 conduites. Appeloms cette pression $P_2$. Alors les 3 dernières équations contiennent 4 inconnues plus les 3 facteurs de friction. Mais ces trois facteurs de friction dépendent de la vitesse donc on a bien 3 équations à 4 inconnues. Ajoutons l'équation de conservation de la masse, équation (1) au système complet:

$$
  \boxed{  w_1 -(w_{21}+w_{22}+w_{23}) = 0} 
  (21)$$

et on peut alors solutionner. Utilisons python pour faire les calculs:

```python
#
# Méthode de solution numérique utilisant le solver de scipy.
#
def f4(y):
#
# Équations
#
  v21,v22,v23,p=y
  eq1=p-L21/D21*v21**2*0.0791*(mu/(rho*v21*D21))**0.25
  eq2=p-L22/D22*v22**2*0.0791*(mu/(rho*v22*D22))**0.25
  eq3=p-L23/D23*v23**2*0.0791*(mu/(rho*v23*D23))**0.25
  eq4=v21*A21+v22*A22+v23*A23-v1*A1
  return (eq1,eq2,eq3,eq4)
#
# Importer l'outil de recherche des zéros d'équations non-linéaires
from scipy.optimize import fsolve
import math
#
# Données numériques et calculs préliminaires
#
D1=0.05
D21=0.04
D22=0.03
D23=0.02
L21=40
L22=30
L23=20
L1=10
pi=math.pi
A1=pi*D1**2/4
A21=pi*D21**2/4
A22=pi*D22**2/4
A23=pi*D23**2/4
q1=300/60/1000             # m3/sec
v1=q1/A1
rho=1000
mu=0.001
#
# Solution
#
v1,v2,v3,p=fsolve(f4,(1,1,1,1))
#
print(v1,v2,v3,p)
print(' Est-ce que le range de Reynolds est ok? ')
print(' Re21, Re22, Re23 = ',rho*v1*D21/mu,rho*v2*D22/mu,rho*v3*D23/mu)
Q21=v1*A21*60000               # l/min
Q22=v2*A22*60000
Q23=v3*A23*60000
Q=Q21+Q22+Q23
print(' Le débit de ',Q,' se sépare donc en ',Q21,Q22,Q23)
```
